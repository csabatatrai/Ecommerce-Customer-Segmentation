# 03 Prediktív Modellezés: CLV és Lemorzsolódás (XGBoost)

**Bemenet:** `data/processed/customer_segments.parquet`  
**Kimenet:** `models/xgboost_clv.joblib`

Ebben a fázisban a gépi tanulás felügyelt részére térünk át. Célunk, hogy a korábban kiszámított RFM metrikák, valamint az azonosított klaszter-címkék alapján előrejelezzük a vásárlók jövőbeli viselkedését (pl. Lemorzsolódás / Churn).

**Szakmai megjegyzés az Adatszivárgásról (Data Leakage) és a Klaszterezésről:**
A `02_customer_segmentation` notebookban generált K-means klasztereket itt bemeneti változóként (feature) használjuk. Fontos azonban megjegyezni, hogy a keresztvalidáció (Cross-Validation) során a klaszterezést nem szabad előre, a teljes adathalmazon elvégezni, mert az adatszivárgáshoz vezetne. Ennek elkerülése érdekében a végső modell kiértékelésénél a `scikit-learn` `Pipeline` modulját alkalmazzuk, amely biztosítja, hogy a K-means algoritmus minden CV-fold esetében szigorúan csak a tréning adatokon tanuljon.
